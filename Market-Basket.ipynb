### Part II: Market Basket Justification


#### B1. Explanation of Market Basket

<p/>

Market Basket Analysis is one of the essential techniques used by large retailers to uncover associations between items. It works by looking for combinations of things that frequently occur in transactions. Association Rules are widely used to analyze retail basket or transaction data. Interestingness measures identify strong rules discovered in transaction data. Support indicates how frequently the item set appears in the data set. Expected confidence is the probability of occurrence of the consequent given the antecedent. The Lift tells us how much better a rule is at predicting the result. (Shah 2022)

<p/>

Market basket analysis is a machine learning tool. Association Rules are also fundamental concepts of Market basket analysis using machine learning. Rules do not extract an individual's preference; they find relationships between a set of elements. Relationships are what makes them different from collaborative filtering. Collaborative filtering reverts all transactions associated with a user ID to identify likenesses between user preferences. User ID helps recommend items on e-commerce websites, songs on Spotify, etc. Rules do not collect user transactions to determine relationships. The support measure gives the frequency an itemset is in transactions. Many transactions will have Bread on the cart with Butter, but Bread with Shampoo? Maybe not so many. Support is the fraction of total transactions the itemset occurs in. The value of support will have, the identifed rules to consider for analysis; a rule to consider only the itemsets that occur at least 50 times out of 10,000 transactions, for example. If an itemset happens to have shallow support, the Market Basket cannot make an association; there is not enough information about the relationship between its items. The confidence measures the likeliness of future item results on a cart, given that the cart already has earlier transacted things. (Garg 2018)

<p/>

We can say by the transacted knowledge that a high confidence rule will exist in the itemset; Captain Crunch and Milk. Since confidence is the conditional probability of occurrence, confidence for an association rule with a persistent consequence will always be high. Let us consider the confidence for Butter and Bread? What is the conditional probability Butter will also have Bread? Very high because the item set is a persistent consequence. Lift measures the frequency of an itemset; while calculating the confidence of Y given X., Supermarkets will have thousands of different products in-store; so finding lift values for each item will get computationally expensive. How do you come up with a set of most important association rules to be considered? The Apriori algorithm. (Garg 2018)

<p/>

Since transactional datasets can hypothetically be enormous, both in terms of the number of transactions and in the number of items purchased, computational cost interrogating large datasets is high, but the Apriori algorithm generates association rules from the transactional data because it reduces the number of itemsets to evaluate. The algorithm utilizes a prior belief about the properties of frequent itemsets called the Apriori Property, which states that all subsets of frequent itemsets must also be frequent. (Junaid 2019)

#### C2. Code Execution

Step 1. Convert the dataset to list of transactions

```python
# dataframe lists (Korstanje 2021)
transactions = []
for i in range(0, 7500):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

Step 2. Run the apriori algorithm.

```python
# run the apriori algorithm (Korstanje 2021)

from efficient_apriori import apriori 

# our min support is 10, but it has to be expressed as a percentage for efficient-apriori
min_support = 20/len(transactions) 

# min confidence allows you to delete rules with low confidence.
# For now set min_confidence = 0 to obtain all the rules
min_confidence = 0.75
itemsets, rules = apriori(transactions, min_support=min_support, min_confidence=min_confidence)
```

#### C3. Association Rules Table

##### Values for the support, lift, and confidence below

```python
# Turning list into a new dataframe (python - Apriori rule to pandas dataframe n.d.)
data = [dict(**rule.__dict__, confidence=rule.confidence, support=rule.support, lift=rule.lift, conviction=rule.conviction)
            for rule in rules]
rules_df = pd.DataFrame(data)
rules_df.sort_values(by ='lift', ascending = False, inplace = True)
rules_df.head(10)
```

### Part IV: Data Summary and Implications



#### D1. Significance of Support, Lift, and Confidence Summary

#### D-1a. Significance of Support

Since the Apriori algorithm is foundational to basket analysis, which is the study of a shoppers basket, the goal is to find products that are often associated together or frequent itemsets; Frequent Itemset Mining. Support identifies the number of transactions an item occurs in, and our current analysis lists all possible combinations. Therefore, it is a common step to set a threshold to raise the minimum occurrence to a more statistically significant number. For this study, the support threshold is 20 with <code>min_support = 20/len(transactions)</code> in Step 2 above. (Korstanje 2021)

#### D-1b. Significance of Confidence

<p/>

Now that the frequent itemsets are known it is time to make them into Association Rules written in the format <code> product X => product Y</code>. Association Rules mean that product X directly relates to product Y in a shopping cart.  Confidence gives a percentage of correctness that items are associated together; 100% means X and Y are always together and 50% means they only occure together half the time. For this study the confidence score is set to a minimum of 75% (a statistically significant level) with <code> min_confidence = 0.75 </code>. The result is a dataset of 1739 records. (Korstanje 2021)

#### D-1c. Significance of Lift

<p/>

Lift now determines the strength of association between the rule and the products found in the basket. Lift compares the improving associations across the dataset. For example, product X  of 10% the cases that improve with a confidence of 75% improves by 7.5. Otherwise, if Lift is one, the products are independent. When the score is higher than one, Lift explains how strongly the right-side object is dependent on the left-side object. (Korstanje 2021)

<p/>

#### D2. Practical Significance of Findings

<p/>

An accuracy in market basket analysis is a decision-based approach to knowing why you prefer one metric of either support, confidence, or lift over the remaining two. All three have validity, but determining why one works over the others in a specific use case is left to the researcher. It is difficult to state that one rule has more merit than another. Conviction may help disambiguate the significance of one item in a rule versus another. A conviction of one means there is no association, like Lift, but anything more significant than one indicates a relationship and direction. Since Lift is the same for X or Y in an itemset, it has no directionality of association. A conviction can be different in X or Y, so a dominant directionality of the relationship is assumed. Practical significance will need to determine what value any of the assumption rules have in a real-world application rather than statical significance to deploy a market basket analysis in a helpful way and obtain a business solution. (Korstanje 2021) (Dobilas 2022)
